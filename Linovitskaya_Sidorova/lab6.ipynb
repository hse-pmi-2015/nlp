```python
import os
from sklearn.datasets import fetch_20newsgroups
import spacy
from spacy import displacy
import pandas as pd
import time
import xml.etree.cElementTree as ET
```

```python
PIPELINE = ['tagger', 'parser', 'ner']
```

```python
tree = ET.ElementTree(file="laba6.xml")
root = tree.getroot()
```

```python
articles=[]
catalog=root[0][0]
for i in range(50):
    articles.append(catalog[i].text)
catalog=root[0][1]
for i in range(50,100):
    articles.append(catalog[i-50].text)
catalog=root[0][2]
for i in range(100,150):
    articles.append(catalog[i-100].text)
catalog=root[0][3]
for i in range(150,200):
    articles.append(catalog[i-150].text)
articles[50]=articles[49]
```

```python
articles[50]
```

```python
articles = [t.replace('\n', ' ') for t in articles]
```

```python
articles
```

```python
nlp = spacy.load('C:/Users/root/Desktop/ЛИНГВИСТИКА/spacy-ru-master/ru2')

for name in PIPELINE:
    component = nlp.create_pipe(name)
```

```python
nlp.pipeline
```

```python
%%time

docs = [nlp(txt) for txt in articles[:50]]
```

```python
%%time

docs = [doc for doc in nlp.pipe(articles[:50], n_threads=6)]
```

## POS

```python
df_pos = pd.DataFrame(columns=['Text', 'Lemma', 'POS', 'TAG', 'DEP', 'Shape', 'Stop Word'])

for token in docs[10]:
    df_pos = df_pos.append({'Text': token.text, 'Lemma': token.lemma_, 'POS': token.pos_, 'TAG': token.tag_, 
                            'DEP': token.dep_, 'Shape': token.shape_, 'Stop Word': token.is_stop}, ignore_index=True)
    
df_pos.head(1000000)
```

## NER

### [annotation types](https://spacy.io/usage/linguistic-features#entity-types)

```python
for doc in docs:
    displacy.render(doc, style='ent', jupyter=True)
    print('=============================================================================================================')
```

```python
df_ent = pd.DataFrame(columns=['Text', 'Start Char', 'End Char', 'Label'])

for ent in docs[10].ents:
    df_ent = df_ent.append({'Text': ent.text, 'Start Char': ent.start_char, 'End Char': ent.end_char, 'Label': ent.label_},
                           ignore_index=True)
    
df_ent.head(10000000)
```

## Dependency Parsing

```python
for sent in docs[10].sents:
    print(sent)
    tmp_doc = sent.as_doc()
    tmp_doc.user_data['title'] = sent.text
    displacy.render(tmp_doc, style='dep', jupyter=True, options={'compact': True})
```
