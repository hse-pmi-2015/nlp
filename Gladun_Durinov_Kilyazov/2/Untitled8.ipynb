```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import gensim
```

```python
data = pd.read_csv('~/Downloads/go_V.csv', sep = ';')
```

```python
data.dropna(inplace = True)
```

```python
x = data.text
y = data.topic
```

```python
X = []
for txt in x:
    kek = txt.split(' ')
    X.append(kek)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.7)
```

```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
documents = [TaggedDocument(x_train[i][:], y_train.values[i]) for i in range(len(x_train))]
model = Doc2Vec(documents, vector_size=20, window=3, min_count=2, workers=4)
```

```python
docs = [TaggedDocument(x_test[i][:], y_test.values[i]) for i in range(len(x_test))]
```

```python
docs[0].words
```

```python
vectorised_text = [model.infer_vector(text.words) for text in docs]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(vectorised_text, y_test, test_size = 0.3)
```

```python
from sklearn import svm
clf = svm.SVC(kernel = 'linear')
clf.fit(X_train, Y_train)  
#SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
#    decision_function_shape='ovr', degree=3, gamma='scale', kernel='rbf',
#    max_iter=-1, probability=False, random_state=None, shrinking=True,
#    tol=0.001, verbose=False)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score( Y_test, clf.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rlf = RandomForestClassifier()
rlf.fit(X_train, Y_train) 
```

```python
print(accuracy_score( Y_test, rlf.predict(X_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
klf = KNeighborsClassifier()
klf.fit(X_train, Y_train) 
```

```python
print(accuracy_score( Y_test, klf.predict(X_test)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
xglf = GradientBoostingClassifier()
score = xglf.fit(X_train, Y_train) 
```

```python
print(accuracy_score( Y_test, xglf.predict(X_test)))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
eclf = VotingClassifier(estimators=[('svc', clf), ('rf', rlf), ('knn', klf), ('xgb', xglf)], voting='hard')
eclf.fit(X_train, Y_train) 
```

```python
print(accuracy_score( Y_test, eclf.predict(X_test)))
```
