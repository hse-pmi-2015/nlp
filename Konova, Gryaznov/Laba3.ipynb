```python
from collections import Counter
import pandas as pd
import math
```

```python
data = pd.read_csv("laba2.csv", header=[0, 1], index_col=0, encoding='utf-8')
```

```python
def tf_idf(corpus):
    def tf(text):
        tf_val = Counter(text)
        for token in tf_val:
            tf_val[token] = tf_val[token] / float(len(text))
            
        return tf_val
    
    def idf(corpus_text):
        idf_val = {}
        tokens = sorted(set([token for text in corpus_text for token in text]))
        for token in tokens:
            idf_val[token] = math.log10(len(corpus_text) /
                                        float(sum([1 for text in corpus_text if token in text])))
        
        return idf_val
    
    corpus_text = []
    
    for category in corpus.columns.levels[0]:
        for i, item in enumerate(corpus[category]['Text']):
            corpus_text.append(item.split())
    
    corpus_val = []
    
    idf_val = idf(corpus_text)
    
    for text in corpus_text:
        tf_idf_val = {}
        tf_val = tf(text)
        
        for word in tf_val:
            tf_idf_val[word] = tf_val[word] * idf_val[word]
            
        corpus_val.append(tf_idf_val)
    
    return corpus_val
```

```python
tf_idf_val = tf_idf(data)
```
