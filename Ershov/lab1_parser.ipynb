```python
def scraper(categoryName):
    
    category = []
    page = 1;
    base_url = 'https://www.mskagency.ru'
    url = 'https://www.mskagency.ru/?type=text&rubric='+str(categoryName)+'&page='+str(page)
    
    while(page <= 4):
        print('URL=' + url)
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'lxml')
        news_list = soup.find('div',{"class","news-list"})
        lis = news_list.findChildren("li" , recursive=True)
        
        links = []
        for i in lis:
            index = str(i).find("href=")
            tmp = str(str(i)[(index+6):])
            indexOfEnd = tmp.find('\"')
            link = tmp[:indexOfEnd]
            links.append(link)
    
        for link in links:
            print(base_url+link)
            r = requests.get(base_url+link)
            tmp_soup = BeautifulSoup(r.text, 'lxml')
            textA = copy.deepcopy(str(tmp_soup.find_all('div',{"class","Text"})))
            category.append(textA)    
            
        page = page + 1
        url = 'https://www.mskagency.ru/?type=text&rubric='+str(categoryName)+'&page='+str(page)
        
    return category
```

```python
text_shit = dict()
categories = ['14','3','5','7'] # Технологии, политика, общество, экономика
for category in categories:
    text_shit[category] = scraper(category)
```
