```python
import pandas as pd
import numpy as np
import re
import string
import stop_words
```

```python
FILE_UNIGRAMS = '1grams-3.txt'
N_WORDS = 192689044
TOL = 1e-20
```

```python
frequencies = {}

with open(FILE_UNIGRAMS, 'r') as f:
    for line in f:
        split = line.lower().strip().split('\t')
        try:
            freq, word = split
        except:
            print(split)
        frequencies[word] = float(freq) / N_WORDS
```

```python
data = pd.read_csv("laba1.csv", header=[0, 1], index_col=0, encoding='utf-8')
data
```

```python
EXCLUDE_SYMBOLS_STR = u''.join(['№', '«', 'ђ', '°', '±', '‚', 'ћ', '‰', '…', '»', 'ѓ', 'µ', '·', 'ґ', 'њ', 'ї', 'џ', 'є', '‹',
                                '‡', '†', '¶', 'ќ', '€', '“', 'ў', '§', '„', '”', '\ufeff', '’', 'љ', '›', '•', '—', '‘', 
                                '\x7f', '\xad', '¤', '\xa0', '¯ツ¯', chr(8211), chr(9583), chr(9633)])
```

```python
regex_punct = re.compile('[%s]' % re.escape(string.punctuation))
regex_dig = re.compile('[%s]' % re.escape(string.digits))
regex_symb = re.compile('[%s]' % re.escape(EXCLUDE_SYMBOLS_STR))
regex_struct = re.compile('[%s]' % string.printable + string.whitespace)
emoji_pattern = re.compile("["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+")
```

```python
for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        for i, item in enumerate(data[category][key]):
            item = emoji_pattern.sub(' ', item)
            item = item.strip().strip('\t').replace('\n', '')
            item = regex_punct.sub(' ', item)
            item = regex_dig.sub(' ', item)
            item = regex_symb.sub(' ', item)
            item = regex_struct.sub(' ', item)
            tokens = []
            sp = item.lower().split()
            for word in sp:
                if word not in stop_words.get_stop_words('ru'):
                    tokens.append(word)
            data.loc[i, (category, key)] = tokens
```

```python
docs_ = []
for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        for i, item in enumerate(data[category][key]):
            docs_.append(item)
```

```python
docs_
```

```python
perp = []

for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        tmp = []
        for i, item in enumerate(data[category][key]):
            frequencies_empirical = {}

            for word in item:
                if frequencies_empirical.get(word, None) is None:
                    frequencies_empirical[word] = 1
                else:
                    frequencies_empirical[word] += 1
            perp.append(frequencies_empirical)
```

```python
perp
```

```python
def perplexity(do, frequencies_empirical):
    out_of_dict = []

    tmp_sum_docs = 0.0
    N = 0.0
    docs = np.array(do)
    for i, doc in enumerate(docs):
        tmp_sum_words = 0.0
        for word in doc:
            freq = frequencies.get(word, TOL)
            freq_empirical = frequencies_empirical[i].get(word, 0.0)
            N += freq_empirical
            if freq == TOL:
                out_of_dict.append(word)
            tmp_sum_words += freq_empirical * np.log(freq)
        tmp_sum_docs += tmp_sum_words

    return np.exp(-tmp_sum_docs / N), out_of_dict
```

```python
perp, ood = perplexity(docs_, perp)

print(perp)
```
