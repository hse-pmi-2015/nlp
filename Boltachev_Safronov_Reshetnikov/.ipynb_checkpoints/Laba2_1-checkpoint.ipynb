# Предобработка

```python
EXCLUDE_SYMBOLS_STR = u''.join(['№', '«', 'ђ', '°', '±', '‚', 'ћ', '‰', '…', '»', 'ѓ', 'µ', '·', 'ґ', 'њ', 'ї', 'џ', 'є', '‹',
                                '‡', '†', '¶', 'ќ', '€', '“', 'ў', '§', '„', '”', '\ufeff', '’', 'љ', '›', '•', '—', '‘', 
                                '\x7f', '\xad', '¤', '\xa0'])
```

```python
regex_punct = re.compile('[%s]' % re.escape(string.punctuation))
regex_dig = re.compile('[%s]' % re.escape(string.digits))
regex_symb = re.compile('[%s]' % re.escape(EXCLUDE_SYMBOLS_STR))
regex_struct = re.compile('[%s]' % string.printable + string.whitespace)
emoji_pattern = re.compile("["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+")
```

```python
for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        for i, item in enumerate(data[category][key]):
            item = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', u'', item)
            itme = re.sub(' +' , ' ', item)
            item = item.strip().strip('\t').replace('\n', u'')
            item = regex_punct.sub('', item)
            item = regex_dig.sub('', item)
            item = regex_symb.sub(' ', item)
            item = regex_struct.sub('', item)
            item = re.sub(' +' , ' ', item.strip())
            item = emoji_pattern.sub('', item)
            data.loc[i, (category, key)] = item
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        for i, item in enumerate(data[category][key]):
            tmp_tokens = []
            sp = item.lower().split()
            for word in sp:
                if word not in stop_words.get_stop_words('ru'):
                    tmp_tokens.append(word)
            data.loc[i, (category, key)] = tmp_tokens
```

```python
stemmer = RussianStemmer()
```

```python
for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        for i, tokens in enumerate(data[category][key]):
            tmp = [stemmer.stem(word) for word in tokens]
            data.loc[i, (category, key)] = tmp
```

# Составление словаря

```python
import itertools
import codecs
```

```python
dictionaries = {}

common_set = set()

for category in data.columns.levels[0]:
    title_set = set(itertools.chain.from_iterable(data[category]['Title']))
    text_set = set(itertools.chain.from_iterable(data[category]['Text']))
    common_set = common_set.union(title_set)
    common_set = common_set.union(text_set)
    dictionaries[category] = {val: i for i, val in enumerate(sorted(title_set.union(text_set)))}

dictionaries[u'common'] = {val: i for i, val in enumerate(sorted(common_set))}
```

```python
for key, item in dictionaries.items():
    with codecs.open("{}.dict".format(key), "w", "utf-8") as f:
        for word, val in item.items():
            f.write(u"{} -> {}\n".format(word, val))
```

# Замена слов на индексы

```python
for category in data.columns.levels[0]:
    for key in data.columns.levels[1]:
        for i, item in enumerate(data[category][key]):
            data.loc[i, (category, key)] = u' '.join([str(dictionaries[u'common'][token])
                                                      for token in item])
```
