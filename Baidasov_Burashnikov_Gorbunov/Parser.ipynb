```python
def scraper(category,n=200):
    culture = []
    base_url = 'https://www.b-port.com'
    url = 'https://b-port.com/' + category + '.html'
    #r = requests.get(url)
    while(len(culture)<n):
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'lxml')
        events = soup.find_all('div',{"class","block news_line_text"})
        links = []
        for i in events:
            tmp = str(list(i)[1])[str(list(i)[1]).find("href")+6:]
            tmp = tmp[:tmp.find('"')]
            links.append(copy.deepcopy(tmp))
        for link in links:
            r = requests.get(base_url+link)
            tmp_soup = BeautifulSoup(r.text, 'lxml')
            culture.append(copy.deepcopy(str(tmp_soup.find_all('div',{"class","itemFullText"}))))
        next_ = str(soup.findAll("div",{'class','Pagination'})).partition('Следующая')[0].rpartition('<')[2].split('"')[1]
        url = base_url + next_
    return culture
```

```python
text_shit = dict()
categories = ['culture','auto','crime','education']
for category in categories:
    text_shit[category] = scraper(category)
```
